```python
counter=0#count number of links

while (1):#terminates when next buttton is unclickable
    elements = driver.find_elements("css selector", ".listing-item")#find elements in web page of "<div class=".listings-container">"
    for element in elements:
        href = element.find_element("css selector", "a").get_attribute('href')#find anchor tags in selected class and extract href attribute
        print( href)
        house_links.append(href)#appends to the list
        counter=counter+1
    next_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//a[contains(text(), 'Older posts')]")))
    next_button.click() 
    #6.12.16.21.23.24.30.31.32.35.43.47.51.64.72.74.99.113
```

```python
city=[]
location=[]
for address in df_homedata['address']:
    if address is not None:
        parts = address.split(',')
        if len(parts) > 1:
            cit = parts[-1].strip()
            city.append(cit)
            address1 = address.replace(cit, '').strip(',')
            location.append(address1)
        else:
            city.append(None)
            location.append(None)
    else:
        city.append(None)
        location.append(None)

df_homedata['district'] = city
df_homedata['location'] = location
df_homedata= df_homedata.drop('address', axis=1)
df_homedata.to_csv('99Aana.csv')
```
