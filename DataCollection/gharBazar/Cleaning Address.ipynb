```python
# Assuming you have a DataFrame named 'df' containing the data
df['LOCATION'] = df['LOCATION'].apply(lambda x: x[:-1] if isinstance(x, list) else x)

# Printing the updated 'LOCATION' column
print(df['LOCATION'])
```

```python
# Assuming you have a DataFrame named 'df' containing the data
df['LOCATION'] = df['LOCATION'].apply(lambda x: x[1:] if isinstance(x, list) and len(x) == 2 else x)

# Printing the updated 'LOCATION' column
print(df['LOCATION'])
```
