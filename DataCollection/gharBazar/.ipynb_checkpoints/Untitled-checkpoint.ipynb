```python
import os, time, csv, re
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.chrome.options import Options
```

```python
# Create ChromeOptions object
options = Options()

# Disable notifications
options.add_argument("--disable-notifications")
path = 'C:/Users/Abiral/Desktop/Scraping/chromedriver_win32/chromedriver.exe'
#loading Driver
service = Service(executable_path=path)
driver = webdriver.Chrome(options=options)
```

```python
# Load the CSV file and extract the URLs
with open('gharbazar_home_urls_duplicate_removed.csv') as csv_file:
    csv_reader = csv.reader(csv_file)
    urls = [row[0] for row in csv_reader]
```

```python
urls.remove(urls[0])
```

```python
driver.get(urls[0])
```

```python
location_container = driver.find_element(by='xpath',value='//*[@id="property_details"]/div/div/div[2]/div[2]/div/ul')
location_child = location_container.find_elements(by='xpath',value='./li')
for child in location_child:
#     print(child.text)
    text = child.text.split(':')
    key = text[0].strip()
    value = text[1].strip() if len(text) > 1 else ''
    print(key , value)
```

```python
detail_container = driver.find_element(by='xpath',value='//*[@id="property_details"]/div/div/div[2]/div[1]/div/ul')
```

```python
child_elements = detail_container.find_elements(by='xpath',value='./li')
```

```python
direction_container = driver.find_element(by='xpath',value='//*[@id="property_details"]/div/div/div[2]/div[1]/div/ul[2]')
direction_child = direction_container.find_elements(by='xpath',value='./li')
for i in direction_child:
    print(i.text)
```

```python
data_list = []
start_time = time.time()
for url in urls: 
    
    driver.get(url)
    data_dict = {}

    price = driver.find_element(by='xpath', value='//*[@id="property-image"]/div/div[3]/div[1]/h2')
    data_dict['price'] = price.text

    detail_container = driver.find_element(by='xpath', value='//*[@id="property_details"]/div/div/div[2]/div[1]/div/ul')
    child_elements = detail_container.find_elements(by='xpath', value='./li')

    for child in child_elements:
        text = child.text.split(':')
        key = text[0].strip()
        value = text[1].strip() if len(text) > 1 else ''
        data_dict[key] = value
        print(key + "=>" +value)

    location_container = driver.find_element(by='xpath',value='//*[@id="property_details"]/div/div/div[2]/div[2]/div/ul')
    location_child = location_container.find_elements(by='xpath',value='./li')

    for child in location_child:
        text = child.text.split(':')
        key = text[0].strip()
        value = text[1].strip() if len(text) > 1 else ''
        data_dict[key] = value

    direction_container = driver.find_element(by='xpath', value='//*[@id="property_details"]/div/div/div[2]/div[1]/div/ul[2]')
    direction_child = direction_container.find_elements(by='xpath', value='./li')

    for child in direction_child:
        text = child.text.split(':')
        key = text[0].strip()
        value = text[1].strip() if len(text) > 1 else ''
        data_dict[key] = value
    data_dict['url'] = url
    print(data_dict)
    data_list.append(data_dict)
end_time = time.time()
print(data_list)
```

```python
df = pd.DataFrame(data_list)
df.to_csv('dataset_gharbazars.csv',index=False)
```
