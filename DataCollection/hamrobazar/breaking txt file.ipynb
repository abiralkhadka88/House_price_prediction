```python
import pandas as pd

# Load the URLs from the text file
with open('new_hamrobazar.txt') as file:
    urls = file.readlines()
    urls = [url.strip() for url in urls]

# Create a DataFrame with a single column 'URL'
df = pd.DataFrame({'URL': urls})

# Display the DataFrame
print(df)
```

```python
df.tail(50)
```

```python
contains_https = df['URL'].str.contains('https')
print(contains_https)
```

```python
df_https = df[contains_https]
print(df_https)
```

```python
df_https.to_csv('hamrobazar_links.csv')
```

```python
# Calculate the number of rows in the DataFrame
num_rows = len(df_https)

# Divide the DataFrame into four equal portions
portion_size = num_rows // 4

# Split the DataFrame into four sub-DataFrames
df10 = df_https[:portion_size]
df20 = df_https[portion_size:2*portion_size]
df30 = df_https[2*portion_size:3*portion_size]
df40 = df_https[3*portion_size:]

# Save each sub-DataFrame to a separate CSV file
df10.to_csv('sub_dataframe1.csv', index=False)
df20.to_csv('sub_dataframe2.csv', index=False)
df30.to_csv('sub_dataframe3.csv', index=False)
df40.to_csv('sub_dataframe4.csv', index=False)
```
