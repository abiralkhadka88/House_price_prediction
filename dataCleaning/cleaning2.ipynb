```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaning_data.csv')
df.shape
```

```python
df['land area'].unique()
```

```python
# Define the conversion functions
def convert_to_aana_sqft(value):
    numeric_part = value.split(' ')[0]
    converted_value = float(numeric_part) / 342.25  # 1 aana = 342.25 sq. ft
    return converted_value

def convert_to_aana_kattha(value):
    numeric_part = value.split(' ')[0]
    converted_value = float(numeric_part) * 20  # 1 kattha = 20 aana
    return converted_value

# Create a copy of the DataFrame
df4 = df.copy()

# Iterate over each value in the 'land area' column
for i, value in enumerate(df4['land area']):
    if 'sq. ft' in value:
        converted_value = convert_to_aana_sqft(value)
        df4.loc[i, 'land area'] = converted_value
        print(f"Converted {value} sq. ft to {converted_value} aana")
    elif 'kattha' in value:
        converted_value = convert_to_aana_kattha(value)
        df4.loc[i, 'land area'] = converted_value
        print(f"Converted {value} kattha to {converted_value} aana")

# Export the modified DataFrame to a CSV file
df4.to_csv('output.csv', index=False)
```
