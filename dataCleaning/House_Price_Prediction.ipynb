```python
def convert_price(df3):
    if df3['price'].str.contains("Cr"):
        df3['price'] = df3['price'].str.replace('Cr', '')  # Remove 'Cr' from values
        df3['price'] = df3['price'].astype(float)
        df3['price'] = df3['price'] * 10000000  # Convert 'Cr' to Nepali rupees
    
    if df3['price'].str.contains("Lac"):
        df3['price'] = df3['price'].str.replace('Lac', '') # Remove 'Lac' from values
        df3['price'] = df3['price'].astype(float)
        df3['price'] = df3['price'] * 100000 # Convert 'Lac' to Nepali rupees
    
    
    return df3
```
