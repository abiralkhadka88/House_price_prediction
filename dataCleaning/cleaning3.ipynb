```python
import matplotlib.pyplot as plt

# Convert 'district' column to string format
df['district'] = df['district'].astype(str)

# Plotting views column
plt.figure(figsize=(10, 6))
plt.bar(df['district'], df['views'])
plt.xlabel('District')
plt.ylabel('Views')
plt.title('Property Views by District')
plt.xticks(rotation=90)
plt.show()
```

```python
import re
import numpy as np

# Remove non-numeric characters from 'parking' column
df['parking'] = df['parking'].apply(lambda x: re.sub('[^0-9]', '', str(x)) if pd.notnull(x) else x)

# Convert numeric strings to integers and sum the individual digits
df['parking'] = df['parking'].apply(lambda x: sum(map(int, str(x))) if pd.notnull(x) else np.nan)

# Print the updated DataFrame
print(df['parking'])
```
