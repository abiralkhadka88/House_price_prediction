```python
has_dash_or_slash['road access'] = has_dash_or_slash['road access'].str.replace('Feet','')
has_dash_or_slash['road access'] = has_dash_or_slash['road access'].str.replace('Meter','')
has_dash_or_slash['road access'] = has_dash_or_slash['road access'].str.replace('+','')
```

```python
has_dash_or_slash['road access'] = has_dash_or_slash['road access'].str.replace('/', '-')
has_dash_or_slash['road access'] = has_dash_or_slash['road access'].str.split('-')
```
